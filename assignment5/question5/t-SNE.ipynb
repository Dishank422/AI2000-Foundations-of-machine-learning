```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits

digits = load_digits()
features = digits.data
labels = digits.target
```

```python
model = TSNE(n_components=2, perplexity=30, n_iter=250)
red_dat = model.fit_transform(features) 
sns.scatterplot(x=red_dat[:,0], y=red_dat[:,1]).set_title("number of iterations = 250")
plt.show()
plt.clf()

model = TSNE(n_components=2, perplexity=30, n_iter=1000)
red_dat = model.fit_transform(features) 
sns.scatterplot(x=red_dat[:,0], y=red_dat[:,1]).set_title("number of iterations = 1000")
plt.show()
plt.clf()

model = TSNE(n_components=2, perplexity=30, n_iter=2000)
red_dat = model.fit_transform(features) 
sns.scatterplot(x=red_dat[:,0], y=red_dat[:,1]).set_title("number of iterations = 2000")
plt.show()
plt.clf()
```
